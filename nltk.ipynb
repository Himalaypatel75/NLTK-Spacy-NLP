```python
import nltk
nltk.download('punkt')
```

```python
text = """The Natural Language Toolkit, or more commonly NLTK, is a suite of libraries and programs for symbolic and statistical natural language processing for English written in the Python programming language."""
```

```python
from nltk.tokenize import word_tokenize
tokenized_words = word_tokenize(text)
print(tokenized_words) #for word tokenization
```

```python
from nltk.tokenize import sent_tokenize
print(sent_tokenize(text)) #for sentense tokenization
```

```python
from nltk.probability import FreqDist
freq_dis = FreqDist(tokenized_words)
print(freq_dis.most_common(20))
print(freq_dis.most_common(3)) #this will give most common repeated words.
```

```python
import matplotlib.pyplot as plt
```

```python
freq_dis.plot(30, cumulative=False)
plt.show() #we can check visually how much frequency per text
```

```python
from nltk.corpus import stopwords
stopwords = stopwords.words('english') #english top words.
stopwords
```

```python
#we need to tokonise sentence.

from nltk.tokenize import word_tokenize, sent_tokenize
words = word_tokenize(text)
words
```

```python
without_stop_words = []
for word in tokenized_words:
    if word not in stopwords:
        without_stop_words.append(word)
        
without_stop_words
```

```python
print(len(without_stop_words) - len(words)) #this will show count of words are eliminated/removed
```

```python
from nltk.stem import PorterStemmer , WordNetLemmatizer
lemeatizer = WordNetLemmatizer()
stemmer = PorterStemmer()
```
